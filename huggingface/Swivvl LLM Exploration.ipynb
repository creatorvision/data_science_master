````python
from transformers import AutoModelForCausalLM, AutoTokenizer

device = "cpu"

model = AutoModelForCausalLM.from_pretrained("mistralai/Mistral-7B-Instruct-v0.2")
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-Instruct-v0.2")

# Example input prompt
prompt = prompt

messages = [{"role": "user", "content": prompt}]

# Apply chat template and tokenize messages
encodeds = tokenizer.apply_chat_template(messages, return_tensors="pt")

# Extract attention mask if available
if "attention_mask" in encodeds:
    attention_mask = encodeds["attention_mask"]
else:
    attention_mask = None

model_inputs = encodeds.to(device)
if attention_mask is not None:
    attention_mask = attention_mask.to(device)

model.to(device)

# Generate output
generated_ids = model.generate(model_inputs.input_ids, attention_mask=attention_mask, max_length=1000, do_sample=True)
decoded = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)
print(decoded[0])
````
