# Loading Dataset

```python
import evalml
```

```python
X,y = evalml.demos.load_breast_cancer()
X_train, X_test, y_train, y_test = evalml.preprocessing.split_data(X,y, problem_type="binary", random_seed =42)
```

```python
X_train.head()
```

## Running the AutoML to select the best algorithm

evalml.problem_types.ProblemTypes.all_problem_types

```python
from evalml.automl import AutoMLSearch
automl = AutoMLSearch(X_train=X_train, y_train=y_train, problem_type='binary')
automl.search()
```

```python
automl.rankings
```

### Best Pipeline

```python
automl.best_pipeline
```

```python
best_pipeline = automl.best_pipeline
```

### Detailed description of the best pipeline

```python
automl.describe_pipeline(automl.rankings.iloc[0]["id"])
```

```python
best_pipeline.score(X_test, y_test, objectives=["auc","f1","Precision","Recall"])
```

## Optimizing as per the objective

```python
automl_auc = AutoMLSearch(X_train=X_train, y_train=y_train,
                         problem_type='binary', 
                         objective ='auc',
                         additional_objectives =['f1','Precision'],
                         max_batches = 1,
                         tolerance = 0,
                         optimize_thresholds = True,
                         random_seed =42)
automl_auc.search()
```

```python
automl_auc.rankings
```

```python
best_pipeline_auc = automl_auc.best_pipeline
```

```python
automl_auc.describe_pipeline(automl_auc.rankings.iloc[0]["id"])
```

```python
best_pipeline_auc.score(X_test, y_test, objectives=["auc","f1","Precision","Recall"])
```

```python
best_pipeline_auc.save('model.pkl')
```

### Load the Model and Test

```python
check_model = best_pipeline_auc.load('model.pkl')
```

```python
check_model.predict_proba(X_test)
```
